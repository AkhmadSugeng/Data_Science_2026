#Data Cleaning Housing Dataset

#STEP 1 - Import Library

Mengimpor library yang dibutuhkan untuk proses data cleaning, yaitu pandas untuk manipulasi data, numpy untuk operasi numerik, dan scipy untuk fungsi statistik tambahan.

#STEP 2 - Load Dataset

Dataset housing_dirty.csv dimuat ke dalam DataFrame menggunakan pandas agar dapat diproses lebih lanjut.

```python
df = pd.read_csv('/housing_dirty.csv')

print("Dataset berhasil dimuat")
df.head()
```

# STEP 4 - Menghapus Data Duplikat

Data duplikat dapat menyebabkan hasil analisis menjadi tidak akurat.

Oleh karena itu dilakukan penghapusan data yang memiliki isi baris yang sama.

```python
print("Jumlah data sebelum:", df.shape)

df.drop_duplicates(inplace=True)

print("Jumlah data sesudah:", df.shape)
```

# STEP 5 - Normalisasi String
Normalisasi dilakukan untuk menyamakan format penulisan data string.

```python
df['kota'] = df['kota'].str.strip().str.title()
df['kondisi'] = df['kondisi'].str.strip().str.lower()

df[['kota', 'kondisi']].head()
```

# STEP 6 - Menangani Missing Values

Missing values ditangani menggunakan:

Median → untuk data numerik
Modus → untuk data kategorikal
Metode median dipilih karena lebih stabil terhadap outlier.

```python
df['luas_m2'] = df['luas_m2'].fillna(df['luas_m2'].median())

df['harga_juta'] = df['harga_juta'].fillna(df['harga_juta'].median())

df['kamar'] = df['kamar'].fillna(df['kamar'].mode()[0])

print("Missing values berhasil ditangani")
```

# STEP 7 - Menangani Outlier

Outlier adalah nilai yang terlalu jauh dari distribusi data normal.

Pada praktikum ini digunakan metode IQR (Interquartile Range) untuk membatasi nilai ekstrem.

```python
for col in ['harga_juta', 'luas_m2']:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower, upper)

print("Outlier berhasil ditangani")
```

**Penyesuaian Tahun Bangun**

Dataset memiliki beberapa nilai tahun bangun yang tidak realistis.

Nilai minimum tahun dibatasi menjadi 1950.

```python
df['tahun_bangun'] = df['tahun_bangun'].clip(lower=1950)
```

# STEP 8 - Validasi Data

Validasi dilakukan untuk memastikan:

* Tidak ada missing values
* Tidak ada data duplikat
* Dataset siap digunakan

```python
print("Total Missing Values :", df.isnull().sum().sum())

print("Total Duplikat :", df.duplicated().sum())

print("Shape Dataset :", df.shape)
```

# STEP 9 - Export Dataset

Dataset yang sudah bersih disimpan ke file baru bernama housing_clean.csv.

```python
df.to_csv('housing_clean.csv', index=False)

print("Dataset berhasil disimpan")
```
